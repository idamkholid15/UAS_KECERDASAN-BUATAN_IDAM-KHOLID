```python
import numpy as np
import matplotlib.pyplot as plt

def tugas_sedikit(x):
    return max(0, min(1, (5 - x) / 5))

def tugas_sedang(x):
    return max(0, min((x - 0) / 5, (10 - x) / 5))

def tugas_banyak(x):
    return max(0, min((x - 5) / 5, 1))

def belajar_sedikit(x):
    return max(0, min(1, (3 - x) / 3))

def belajar_sedang(x):
    return max(0, min((x - 0) / 3, (6 - x) / 3))

def belajar_banyak(x):
    return max(0, min((x - 3) / 3, 1))

def tidur_sedikit(x):
    return max(0, min(1, (6 - x) / 6))

def tidur_sedang(x):
    return max(0, min((x - 0) / 6, (8 - x) / 6))

def tidur_banyak(x):
    return max(0, min((x - 6) / 6, 1))

def inferensi(jumlah_tugas, waktu_belajar, waktu_tidur):
    # Menentukan nilai keanggotaan untuk setiap input
    t_sedikit = tugas_sedikit(jumlah_tugas)
    t_sedang = tugas_sedang(jumlah_tugas)
    t_banyak = tugas_banyak(jumlah_tugas)

    b_sedikit = belajar_sedikit(waktu_belajar)
    b_sedang = belajar_sedang(waktu_belajar)
    b_banyak = belajar_banyak(waktu_belajar)

    t_sedikit_tidur = tidur_sedikit(waktu_tidur)
    t_sedang_tidur = tidur_sedang(waktu_tidur)
    t_banyak_tidur = tidur_banyak(waktu_tidur)

    # Aturan Fuzzy 
    stress_rendah = min(t_sedikit, b_banyak, t_banyak_tidur)
    stress_sedang = min(t_sedang, b_sedang, t_sedang_tidur)
    stress_tinggi = min(t_banyak, b_sedikit, t_sedikit_tidur)

    # Defuzzifikasi 
    numerator = (stress_rendah * 25 + stress_sedang * 50 + stress_tinggi * 75)
    denominator = (stress_rendah + stress_sedang + stress_tinggi)
    
    if denominator == 0:
        return 0, "Tidak dapat dihitung" 
    
    tingkat_stres = numerator / denominator

    if tingkat_stres <= 33:
        keterangan = "Rendah"
    elif tingkat_stres <= 66:
        keterangan = "Sedang"
    else:
        keterangan = "Tinggi"

    return tingkat_stres, keterangan

# Input 
jumlah_tugas = 8
waktu_belajar = 5
waktu_tidur = 7

tingkat_stres, keterangan = inferensi(jumlah_tugas, waktu_belajar, waktu_tidur)
print(f"Tingkat Stres: {tingkat_stres:.2f} - Keterangan: {keterangan}")
```
